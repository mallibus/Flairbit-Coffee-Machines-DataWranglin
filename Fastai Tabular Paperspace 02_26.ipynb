```python
# Temp fix 
df = pd.read_csv("Faults_Cleaning_100_D_20_1.csv",parse_dates = ['Target Timestamp'])
```

```python
df.info()
```

```python
df.loc[df['TARGET']>0,'TARGET'] = 1
```

```python
df['TARGET'].value_counts()
```

```python
df['Target Timestamp'].min()
df['Target day'] = (df['Target Timestamp']-df['Target Timestamp'].min()).apply(lambda d: d.days)
```

```python
df.drop(['Unnamed: 0','Target Timestamp','Serial'],axis=1,inplace=True)
```

```python
df['TARGET'] = df['TARGET'].astype(int)
```

```python
df.columns
```

```python
dep_var = 'TARGET'
cat_names = ['Model']
cont_names = ['CLEANING-20D', 'CLEANING-19D', 'CLEANING-18D', 'CLEANING-17D',
       'CLEANING-16D', 'CLEANING-15D', 'CLEANING-14D', 'CLEANING-13D',
       'CLEANING-12D', 'CLEANING-11D', 'CLEANING-10D', 'CLEANING-9D',
       'CLEANING-8D', 'CLEANING-7D', 'CLEANING-6D', 'CLEANING-5D',
       'CLEANING-4D', 'CLEANING-3D', 'CLEANING-2D', 'CLEANING-1D', 'FAULT-20D',
       'FAULT-19D', 'FAULT-18D', 'FAULT-17D', 'FAULT-16D', 'FAULT-15D',
       'FAULT-14D', 'FAULT-13D', 'FAULT-12D', 'FAULT-11D', 'FAULT-10D',
       'FAULT-9D', 'FAULT-8D', 'FAULT-7D', 'FAULT-6D', 'FAULT-5D', 'FAULT-4D',
       'FAULT-3D', 'FAULT-2D', 'FAULT-1D', 'Target day']
procs = [FillMissing, Categorify, Normalize]
```

```python
test = TabularList.from_df(df.iloc[52200:62200].copy(), cat_names=cat_names,cont_names=cont_names)
```

```python
data = (TabularList.from_df(df, cat_names=cat_names, cont_names=cont_names, procs=procs)
                           .split_by_idx(list(range(52200,62200)))
                           .label_from_df(cols=dep_var)
                           .add_test(test)
                           .databunch())
```

```python
data.show_batch(rows=10)
```

```python
learn = tabular_learner(data, layers=[200,100], metrics=accuracy)
```

```python
learn.fit(1, 1e-2)
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit(10, 1e-4)
```
